```python
import yfinance as yf
import pandas as pd
import datetime

# Function to download hourly BTC data in chunks
def download_hourly(symbol="BTC-USD", start="2015-01-01", end="2025-09-16"):
    start_dt = pd.to_datetime(start)
    end_dt = pd.to_datetime(end)
    
    all_data = []

    # Loop in 60-day chunks
    while start_dt < end_dt:
        chunk_end = min(start_dt + pd.Timedelta(days=59), end_dt)
        
        print(f"Fetching {start_dt.date()} to {chunk_end.date()} ...")
        data = yf.download(
            symbol,
            start=start_dt.strftime("%Y-%m-%d"),
            end=chunk_end.strftime("%Y-%m-%d"),
            interval="1h"
        )
        
        if not data.empty:
            all_data.append(data)
        
        start_dt = chunk_end + pd.Timedelta(days=1)

    # Combine all chunks
    if all_data:
        full_data = pd.concat(all_data)
        full_data = full_data[~full_data.index.duplicated(keep="first")]  # remove duplicates
        return full_data
    else:
        return pd.DataFrame()

# Download BTC hourly from 2015 to 2025
btc_hourly = download_hourly("BTC-USD", "2018-01-01", "2025-01-01")

# Save to CSV
btc_hourly.to_csv("data/BTC-USD-2018-hourly_yf.csv")
print("Saved data shape:", btc_hourly.shape)
print(btc_hourly.head())
print(btc_hourly.tail())
```
