```python
import pandas as pd
from binance.client import Client
import datetime

# Public client (no API key needed for historical klines)
client = Client()

def get_binance_data(symbol="BTCUSDT", interval=Client.KLINE_INTERVAL_1HOUR,
                     start="2017-01-01", end="2025-09-16"):
    klines = client.get_historical_klines(symbol, interval, start, end)
    df = pd.DataFrame(klines, columns=[
        "OpenTime", "Open", "High", "Low", "Close", "Volume",
        "CloseTime", "QuoteAssetVolume", "Trades",
        "TakerBuyBaseVolume", "TakerBuyQuoteVolume", "Ignore"
    ])
    df["OpenTime"] = pd.to_datetime(df["OpenTime"], unit="ms")
    df.set_index("OpenTime", inplace=True)
    df = df[["Open", "High", "Low", "Close", "Volume"]].astype(float)
    return df

btc_binance = get_binance_data()
btc_binance.to_csv("data/BTC-USD-Binance-hourly.csv")
print(btc_binance.head(), btc_binance.tail())
```
